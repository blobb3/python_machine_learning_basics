# OLS-Regressionsergebnisse



#### 1. **R-squared (R²)** und **Adjusted R-squared**
   - **R-squared (R²)** gibt an, wie viel Prozent der Varianz der Zielvariable durch das Modell erklärt wird. Ein hoher Wert (nahe 1) zeigt eine gute Anpassung des Modells an die Daten, während ein niedriger Wert (nahe 0) darauf hinweist, dass das Modell die Zielvariable kaum erklären kann.
   - **Adjusted R-squared** ist eine angepasste Version von R², die die Anzahl der Prädiktoren berücksichtigt und Überanpassung bestraft. Adjusted R² ist besonders nützlich bei Modellen mit mehreren Prädiktoren. Auch hier gilt: Ein höherer Wert ist besser.

   **Hinweis**: Werte über 0.5 für R² und Adjusted R² gelten als moderat bis gut, während Werte unter 0.3 meist auf eine schwache Modellanpassung hindeuten.

#### 2. **F-statistic und Prob (F-statistic)**
   - Die **F-statistic** testet, ob das gesamte Modell signifikant ist. Ein höherer Wert zeigt an, dass das Modell besser ist als ein Modell ohne erklärende Variablen.
   - **Prob (F-statistic)** ist der p-Wert für den F-Test und sollte niedrig (unter 0.05) sein, um auf statistische Signifikanz hinzudeuten. Ein hoher p-Wert deutet darauf hin, dass das Modell möglicherweise die Zielvariable nicht ausreichend erklärt.

   **Hinweis**: Ein signifikanter F-Test (p-Wert < 0.05) zeigt, dass das Modell als Ganzes sinnvoll ist.

#### 3. **Koeffizienten (coef)**
   - Die **Koeffizienten** (`coef`) zeigen den Einfluss jeder unabhängigen Variable auf die Zielvariable. Ein positiver Koeffizient bedeutet, dass die Zielvariable steigt, wenn die Prädiktorvariable zunimmt, während ein negativer Koeffizient das Gegenteil bedeutet.
   - Der **Interzept (const)** gibt den geschätzten Wert der Zielvariablen an, wenn alle Prädiktoren null sind. Dieser Wert ist oft weniger relevant für die Interpretation, aber kann nützlich sein, um den Basiswert der Zielvariable zu kennen.

   **Hinweis**: Die Richtung und Größe der Koeffizienten hängen von der Skalierung der Daten ab und müssen im Kontext der Variablen interpretiert werden.

#### 4. **p-Wert (P>|t|)** für die Koeffizienten
   - Der **p-Wert** für jeden Koeffizienten zeigt an, ob dieser statistisch signifikant ist. Ein p-Wert unter 0.05 deutet darauf hin, dass der Prädiktor einen signifikanten Einfluss auf die Zielvariable hat.
   - Koeffizienten mit hohen p-Werten (> 0.05) sind in der Regel nicht signifikant und können in zukünftigen Modellen möglicherweise weggelassen werden.

   **Hinweis**: Werte unter 0.05 gelten als statistisch signifikant und sind für die Modellinterpretation aussagekräftig.

#### 5. **Konfidenzintervalle (0.025, 0.975)**
   - Die Spalten 0.025 und 0.975 geben das **95%-Konfidenzintervall** für die Koeffizienten an. Sie zeigen den Bereich, in dem der wahre Wert des Koeffizienten mit 95%-iger Wahrscheinlichkeit liegt.
   - Ein Konfidenzintervall, das Null enthält, deutet darauf hin, dass der Koeffizient möglicherweise keinen signifikanten Einfluss auf die Zielvariable hat.

#### 6. **AIC und BIC**
   - **AIC (Akaike Information Criterion)** und **BIC (Bayesian Information Criterion)** sind Modellgütemaße, die zur Modellbewertung und -auswahl verwendet werden. Niedrigere Werte deuten auf eine bessere Anpassung und Effizienz des Modells hin, insbesondere beim Vergleich verschiedener Modelle.
   - BIC bestraft zusätzliche Parameter stärker als AIC, daher wird BIC häufig für die Wahl zwischen Modellen bevorzugt.

   **Hinweis**: Bei der Auswahl zwischen Modellen wird das Modell mit dem niedrigeren AIC- oder BIC-Wert bevorzugt.

#### 7. **Diagnosemetriken**
   - **Omnibus** und **Prob(Omnibus)**: Diese Tests prüfen die Normalität der Residuen. Ein niedriger Prob(Omnibus)-Wert deutet darauf hin, dass die Residuen nicht normal verteilt sind.
   - **Durbin-Watson**: Dieser Test prüft die Autokorrelation der Residuen. Werte nahe 2 sind ideal, während Werte deutlich unter 1 oder über 3 auf Autokorrelation hindeuten, was problematisch sein kann.
   - **Jarque-Bera (JB)** und **Prob(JB)**: Diese Tests prüfen ebenfalls die Normalität der Residuen. Ein hoher p-Wert (> 0.05) ist wünschenswert und deutet darauf hin, dass die Residuen normal verteilt sind.
   - **Skew** und **Kurtosis**: Diese Metriken beschreiben die Verteilung der Residuen. Werte nahe 0 (für Skew) und nahe 3 (für Kurtosis) sind wünschenswert.

   **Hinweis**: Die Normalverteilung der Residuen ist eine Annahme für OLS-Modelle. Abweichungen davon können auf Modellprobleme hinweisen.

#### 8. **Condition Number (Cond. No.)**
   - Der **Condition Number** ist ein Indikator für Multikollinearität. Ein hoher Wert (üblicherweise über 30) deutet darauf hin, dass es starke Korrelationen zwischen den Prädiktoren gibt, was die Stabilität des Modells beeinträchtigen kann.

   **Hinweis**: Ein hoher Condition Number kann auf Multikollinearitätsprobleme hindeuten und erfordert möglicherweise eine Anpassung der Prädiktoren.

---

### Zusammenfassung für eine generische Interpretation
- **Hoher R²/Adjusted R²**: Gute Anpassung des Modells.
- **Niedriger p-Wert für F-statistic**: Das Modell ist insgesamt signifikant.
- **Signifikante p-Werte für einzelne Koeffizienten (< 0.05)**: Die Prädiktoren haben einen signifikanten Einfluss auf die Zielvariable.
- **Niedrige AIC/BIC-Werte**: Gutes Modell im Vergleich zu anderen.
- **Normalität der Residuen (Omnibus, Durbin-Watson, Jarque-Bera)**: Unterstützt die Annahmen des OLS-Modells.
- **Niedriger Condition Number (< 30)**: Geringe Multikollinearität und stabile Koeffizienten.

Diese Kennzahlen helfen dir, die Eignung und die Zuverlässigkeit eines OLS-Modells zu bewerten.

### Zusammengefasst: Schnellbeurteilung eines Modells

Um auf einen Blick zu sehen, ob ein Modell "etwas taugt", prüfe die folgenden Punkte:

1. **R² und Adjusted R²**: Werte über 0.5 sind ein gutes Zeichen.
2. **Prob (F-statistic)**: Sollte < 0.05 sein, damit das Modell insgesamt signifikant ist.
3. **p-Werte der Koeffizienten**: Sollten für die wichtigsten Variablen < 0.05 sein.
4. **AIC und BIC**: Niedrigere Werte sind besser (besonders im Vergleich zu alternativen Modellen).
5. **Condition Number**: Sollte nicht über 30 liegen, um Multikollinearitätsprobleme zu vermeiden.

Wenn diese Werte akzeptable Grenzen erfüllen, zeigt das, dass das Modell grundsätzlich brauchbar ist.
