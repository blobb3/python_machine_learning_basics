# Simple Linear Regression: Confidence- and Prediction-Intervals

This notebook demonstrates how to create confidence and prediction intervals for a linear regression model using Python, based on rental price data (Y) and living area (X).

See also: https://lmc2179.github.io/posts/confidence_prediction.html

## Libraries and Settings

```python
# Import libraries
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.formula.api as smf

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

# Show current working directory
print(os.getcwd())
```

## Generate Apartment Data
We will use the provided data representing rental prices (`Y`) and living area (`X`).

Zusammengefasst generiert dieser Abschnitt die benötigten Daten für eine einfache lineare Regression, bei der living_area als unabhängige Variable (Prädiktor) und rental_price als abhängige Variable (Zielgröße) dient. Die Daten werden dann in einem DataFrame gespeichert, sodass sie im nächsten Schritt für das Erstellen eines Regressionsmodells verwendet werden können.

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Generate values for rental_price and living_area
x = np.linspace(45, 160, 25)
y = np.interp(x, [45, 160], [1500, 4900]) + np.random.normal(0, 400, 25)

# Create a new DataFrame with 25 data points
data = pd.DataFrame({
    'rental_price': y,
    'living_area': x
})

df = pd.DataFrame(data)
df.columns = ['rental_price', 'living_area']
df.head()
```

## Fit an OLS Regression Model
We will use `statsmodels` to fit an ordinary least squares (OLS) regression model for rental price as a function of living area.

```python
# Fit OLS regression
model = smf.ols('rental_price ~ living_area', data=df)

# Get and show results
results = model.fit()
print(results.summary())

# Plot data and regression line
plt.scatter(df['living_area'], df['rental_price'])
plt.plot(df['living_area'], 
         results.predict(df['living_area']), 
         color='darkred', 
         linewidth=2)
plt.xlabel('Living Area')
plt.ylabel('Rental Price')
plt.title('Living Area vs Rental Price')

# Add annotation (regression equation with f-string)
plt.text(50, 4500, f"y = {results.params[0]:.2f} + {results.params[1]:.2f}x", 
         fontsize=12, color='black')

plt.show()
```

The output shows a simple linear regression analysis for predicting `rental_price` based on `living_area`. Here’s a breakdown:

1. **Regression Equation**:
   - The regression line equation is \( y = 561.49 + 25.10x \).
   - This means that for each additional square meter of living area, the rental price is expected to increase by approximately 25.10 units, holding other factors constant.

2. **Interpretation of Coefficients**:
   - **Intercept (561.49)**: When the living area is zero, the base rental price would be approximately 561.49. Although in practice, a living area of zero doesn't make sense, this intercept gives the starting point of the model.
   - **Slope (25.10)**: Each additional square meter of living area adds about 25.10 units to the rental price. This is the estimated marginal effect of living area on rental price.

3. **Goodness of Fit (R-squared)**:
   - The \( R^2 \) value is 0.866, which means that approximately 86.6% of the variance in rental price can be explained by the living area. This indicates a strong linear relationship between the variables.

4. **P-values**:
   - Both the intercept and the `living_area` coefficient have low p-values (0.019 for the intercept and almost 0 for `living_area`), indicating that they are statistically significant predictors of rental price at common significance levels (e.g., 0.05).

5. **Regression Plot**:
   - The scatter plot shows the observed data points for rental prices versus living area, with a regression line overlaid.
   - The line fits well through the center of the data points, supporting the strong \( R^2 \) value and the conclusion that there’s a positive linear relationship between living area and rental price.

In summary, this model suggests that `living_area` is a significant predictor of `rental_price` with a strong linear fit, as evidenced by the high \( R^2 \) value and the well-aligned regression line.

## Generate Prediction- and Confidence Intervals

We calculate the prediction- and confidence intervals for the model.

```python
# Generate predictions and confidence intervals
alpha = 0.05
predictions = results.get_prediction(df).summary_frame(alpha)
predictions.head()
```

## Visualizing the Regression Line and Confidence Interval

We now plot the observed data along with the regression line and confidence intervals.

```python
# Plot observed data and regression line
plt.scatter(df['living_area'], 
            df['rental_price'], 
            label='Observed', 
            marker='x', 
            color='black')

# Plot regression line
plt.plot(df['living_area'], 
         predictions['mean'], 
         label='Regression line')

# Plot confidence intervals
plt.fill_between(df['living_area'], 
                 predictions['mean_ci_lower'], 
                 predictions['mean_ci_upper'], 
                 color='blue', 
                 alpha=0.4, 
                 label='95% Confidence Interval')

# Add legend and labels
plt.xlabel('Living Area (m2)')
plt.ylabel('Rental Price (CHF)')
plt.legend()
plt.show()
```

## Visualizing the Regression Line and Prediction Interval

We now plot the observed data along with the regression line, confidence- and prediction intervals.

```python
# Plot observed data and regression line
plt.scatter(df['living_area'], 
            df['rental_price'], 
            label='Observed', 
            marker='x', 
            color='black')

# Plot regression line
plt.plot(df['living_area'], 
         predictions['mean'], 
         label='Regression line')

# Plot prediction interval
plt.fill_between(df['living_area'], 
                 predictions['obs_ci_lower'], 
                 predictions['obs_ci_upper'], 
                 alpha=.5, 
                 label='95% Prediction interval',
                 color='orange')

# Plot confidence interval
plt.fill_between(df['living_area'], 
                 predictions['mean_ci_lower'], 
                 predictions['mean_ci_upper'], 
                 alpha=.4, 
                 label='95% Confidence interval',
                 color='blue')

# Add labels
plt.xlabel('Living area (m2)')
plt.ylabel('Rental price (CHF)')
plt.legend()

# Show plot
plt.show()
```
